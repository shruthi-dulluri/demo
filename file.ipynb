```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_excel('AlibabaSales_lunar_sdx.xlsx',index_col='Date',parse_dates=['Date','Time'])
```

```
type(df.index)
```

```
df['inputDate'] = pd.to_datetime(df.index) 
df['dayOfWeek'] = df['inputDate'].dt.day_name() 
  
```

```
df.head()
```

```
df['inputmonth'] = pd.to_datetime(df.index) 
df['month'] = df['inputDate'].dt.month_name() 
  
```

```
df.head()
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.countplot(x='Product line',data =df,hue ='Gender',ax =ax)
```

men mostly spend on health and beauty and women mostly on fashion accecories

```
fig,ax = plt.subplots(figsize =(12,6))
sns.countplot(x='Product line',data =df,hue ='Customer type',ax =ax)
```

in Branch A there are most products of home and life style sold while branch B and c most of the goods sold are of Food and beverages ,fashion accessories

```
fig,ax = plt.subplots(figsize =(12,6))
sns.countplot(x='Product line',data =df,hue ='Branch',ax =ax)
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='dayOfWeek',y= 'Total',data = df,hue='Product line',ax =ax)
ax.set_title('sales on day of the week based on Type of product')
ax.legend(bbox_to_anchor = (0,-0.10,1,0))
```

Sales of health and beauty products happen mostly on saturdays and female shop mostly of tuesdays and fridays and men mostly in weekend

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='dayOfWeek',y='Total',data = df,hue ='Gender')
ax.set_title('total of each day of week w.r.t gender')
```

generally women buy mostly on wednesdays and fridays but there are outliers

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='dayOfWeek',y='Total',data = df,hue ='Product line')
ax.set_title('total of each day of week w.r.t type of produts')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='dayOfWeek',y='Total',data = df,hue ='City')
ax.set_title('total of each day of week w.r.t city or branch')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

city B has most sales on friday and wednesday and city A and c have sales mostly on tuesdays

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='month',y='Total',data = df,hue ='Gender')
ax.set_title('total for each month w.r.t gender')
```

total is high for male in month of december and very less in month of august

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='month',y='Total',data = df,hue ='Product line')
ax.set_title('total for each month w.r.t to goods sold')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

Home and lifestyle goods are mostly sold in Feb, Fashion accessories in december,sports and travel in september.

```
fig,ax = plt.subplots(figsize =(12,6))
sns.boxplot(x='month',y='Total',data = df,hue ='City')
ax.set_title('total for each month w.r.t branch or city')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.std,ax =ax) # female has more variation in bill
```

```
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean) # avarage
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean,hue='Product line',ax =ax)
ax.set_title('averages of gender for each product category')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean,hue='dayOfWeek',ax = ax)
ax.set_title('average of totals of gender per each day of week')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean,hue='month',ax = ax)
ax.set_title('average of totals of gender per each month')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

generally women tend to spend less in june,men in august and april

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean,hue='City',ax = ax)
ax.set_title('average of totals of gender per city')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

```
fig,ax = plt.subplots(figsize =(12,6))
sns.barplot(x='Gender',y='Total',data =df ,estimator=np.mean,hue='Customer type',ax = ax)
ax.set_title('average of totals of gender per city')
ax.legend(bbox_to_anchor=(0,-0.10,1,0))
```

generally memembers tend to spend more that non-members
